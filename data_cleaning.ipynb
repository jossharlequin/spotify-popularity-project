```python
# Importing dependencies 
import pandas as pd
from pathlib import Path
import csv
import json
```

```python
# Read the spotify data in from data.csv
url = 'https://media.githubusercontent.com/media/jossharlequin/spotify-popularity-project/main/Resources/data.csv'
df = pd.read_csv(url)
```

```python
# Converting ms to seconds and removing columns with nonnumerical values
df['duration_seconds'] = df['duration_ms']/1000
df = df.drop(columns=['artists', 'id', 'name', 'release_date', 'year', 'duration_ms'])
df.head()
```

```python
df.to_csv('Resources/cleaned_spotify_data.csv', index=False)
```

# Webpage Data

```python
# Read spotify data
url = 'https://media.githubusercontent.com/media/jossharlequin/spotify-popularity-project/main/Resources/data.csv'
exp_df = pd.read_csv(url)

# Export data for website
export_data2015 = exp_df.loc[exp_df['year'] == 2015]
export_data2016 = exp_df.loc[exp_df['year'] == 2016]
export_data2017 = exp_df.loc[exp_df['year'] == 2017]
export_data2018 = exp_df.loc[exp_df['year'] == 2018]
export_data2019 = exp_df.loc[exp_df['year'] == 2019]
export_data2020 = exp_df.loc[exp_df['year'] == 2020]
export_data = pd.concat([export_data2015, export_data2016, export_data2017, export_data2018, export_data2019, export_data2020])
export_data = export_data.sort_values(by=['name'])
export_data = export_data.drop_duplicates(subset=['name'])
export_data.to_csv('Resources/export_data.csv')

# Convert export data to JSON
gvn_csvfile = 'Resources/export_data.csv'
gvn_jsonfile = 'Resources/export_data.json'
json_dictionary = {}
with open(gvn_csvfile) as gvn_csvfile:
    csvfiledata = csv.DictReader(gvn_csvfile)
    json_dictionary['data']= []
    for row_data in csvfiledata:
        print(row_data)
        json_dictionary['data'].append(row_data)
with open(gvn_jsonfile, 'w') as gvn_jsonfile:
    gvn_jsonfile.write(json.dumps(json_dictionary, indent=4))
```
